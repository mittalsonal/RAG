```python
### Text Loaders
from langchain_community.document_loaders import TextLoader
```

```python
'''What is encoding?

Text files are actually stored as bytes (numbers).
Encoding tells Python:

“How do I convert these bytes into readable characters?”

✅ What is "utf-8"?

UTF-8 is:

The most common encoding in the world

Supports:

English ✅

Emojis 😄 ✅

Hindi, Arabic, Chinese, etc. ✅

💡 Why you need it

Without encoding, Python guesses — and sometimes fails.

'''

loader = TextLoader("../data/text_files/python_intro.txt", encoding="utf-8")  #encoding="utf-8" 👉 This tells Python how to read the text file
document = loader.load()
print (document)
```
