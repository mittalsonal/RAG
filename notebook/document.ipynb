```python
### Text Loaders
from langchain_community.document_loaders import TextLoader
```

```python
'''What is encoding?

Text files are actually stored as bytes (numbers).
Encoding tells Python:

“How do I convert these bytes into readable characters?”

✅ What is "utf-8"?

UTF-8 is:

The most common encoding in the world

Supports:

English ✅

Emojis 😄 ✅

Hindi, Arabic, Chinese, etc. ✅

💡 Why you need it

Without encoding, Python guesses — and sometimes fails.

'''

loader = TextLoader("../data/text_files/python_intro.txt", encoding="utf-8")  #encoding="utf-8" 👉 This tells Python how to read the text file
document = loader.load()
print (document)
```

```python
### Directory Loader
from langchain_community.document_loaders import DirectoryLoader

## This will load all the text files in the directory and create a list of Document objects
dir_loader = DirectoryLoader(
    "../data/text_files",
    glob="**/*.txt",  # pattern to match all text files in the directory and subdirectories
    loader_cls= TextLoader, ## Loader class to use for loading the files
    loader_kwargs={'encoding': "utf-8"},
    show_progress=False, # show_progress=True will show a progress bar while loading the files
)

documents = dir_loader.load()
documents
```

```python
from langchain_community.document_loaders import PyMuPDFLoader 

# PDF Loaders
dir_loader = DirectoryLoader(
    "../data/pdf",
    glob="**/*.pdf",  
    loader_cls= PyMuPDFLoader, 
    show_progress=False, 
)

pdf_documents = dir_loader.load()
pdf_documents
```

```python
from langchain_community.document_loaders import JSONLoader
from langchain_community.document_loaders import DirectoryLoader

# JSON loader for db folder
db_loader = DirectoryLoader(
    "../data/db",
    glob="**/*.json",
    loader_cls=JSONLoader,
    loader_kwargs={
        "jq_schema": ".",   # pura JSON read karega
        "text_content": False
    }
)

db_documents = db_loader.load()

db_documents
```

```python
import json
import os
from langchain_core.documents import Document

documents = []

folder_path = "../data/db"

for file in os.listdir(folder_path):
    if file.endswith(".json"):
        with open(os.path.join(folder_path, file), "r", encoding="utf-8") as f:
            data = json.load(f)

            documents.append(
                Document(
                    page_content=str(data),
                    metadata={"source": file}
                )
            )

documents
```
